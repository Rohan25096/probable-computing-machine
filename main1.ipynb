```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
data = pd.read_csv('shoe_size.csv',encoding='ISO-8859-1')
data.head()
```

```python
dummies_sex = pd.get_dummies(data.sex).astype(int)
```

```python
merged = pd.concat([data,dummies_sex],axis='columns')
```

```python
final = merged.drop(['sex','m'],axis = 'columns')
```

```python
X = final.drop(['price(£)','shoe_size'],axis = 'columns')
y = final[['price(£)', 'shoe_size']]
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
print("Model Training Score:", model.score(X, y))
```
